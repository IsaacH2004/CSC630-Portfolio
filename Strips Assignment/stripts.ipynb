## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import altair as alt
```

## Obtain Data

source: https://gist.github.com/natejessee/c53b539bb2dce90fd917928c2ec5b6f8

```python
data = pd.read_csv("fastfood_calories.csv")
data
```

## Just the Strips

We have no need for the data that does not involve chicken strips. 

```python
stripsdata = pd.DataFrame()

for row in data.index:
    if "strips" in data.loc[row, "item"] or "Strips" in data.loc[row, "item"]:
        stripsdata = pd.concat([stripsdata, data.loc[row, ["item", "chain", "calories", "total_fat"]]], axis=1)

stripsdata = stripsdata.T # transpose
stripsdata
```

## Plot

```python
# Plot 1: A scatterplot
alt.Chart(
    stripsdata, title = "Fat and Calorie Counts of Chicken Strips at Select Food Chains"
).mark_point(
    point=True,
    size=200
).encode(
    alt.X("calories:Q", axis=alt.Axis(grid=False, title="Item Calories")),
    alt.Y('total_fat:Q', axis=alt.Axis(grid=False, title="Total Fat (grams)")),
    alt.Color("chain:N"), 
    alt.Tooltip(["item", "chain", "calories", "total_fat"])
).properties(
    width=600,
    height=600
)
```

```python
# Plot 2: A Bar plot
alt.Chart(
    stripsdata, title = "Total Fat Content of Chicken Strips at Select Food Chains"
).mark_bar(
    size=200,
    filled = False
).encode(
    alt.X("chain:N", axis=alt.Axis(grid=False, title="Food Chain")),
    alt.Y('total_fat:Q', axis=alt.Axis(grid=False, title="Total Fat (grams)")),
    alt.Color("chain:N"), 
    alt.Tooltip(["item", "chain", "calories", "total_fat"])
).properties(
    width=600,
    height=600
).interactive()

# A bit confusing, but I honestly like it! The graph has a minimaliism to it that I appreciate. 
```

## Conclusions

In general, Sonic's chicken strips have the highest fat and calorie amounts. However, this can be attributed to the fact that Sonic has the largest portion sizes (3, 4, 5 strip options). Chick Fil-A has the greatest number of options (you can get anywhere from 1 to 4 wings as your order), while Dairy Queen only has a single option, that being the 3-strip option.
