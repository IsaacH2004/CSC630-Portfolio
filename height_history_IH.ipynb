## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Creating the Dataset

```python
hdata = pd.Series({1:1, 2:1.5, 3:2, 4:2.333, 
                   5:2.5, 6:2.75, 7:3, 8:3.65, 
                   9:3.960, 10:4.659, 11:5, 12:5.25,
                   13:5.466, 14:5.7, 15:5.71, 16:5.71,
                   17:5.71}) 
```

This is a dataset based on my height over time (rough estimates based on my memory). To see the progression of my height, we must graph this dataset. 

First, however, we should make record of a few important events in my life so that we have the ability to see what influenced my growth: 

```python
pa_entrance = pd.Series({"x":[np.linspace(14,14,50)], "y":[np.linspace(0,6,50)]}) # I enter PA at 14
started_running = pd.Series({"x":[np.linspace(12,12,50)], "y":[np.linspace(0,6,50)]}) # began long-distance running
middle_schl = pd.Series({"x":[np.linspace(10,10,50)], "y":[np.linspace(0,6,50)]}) # entered middle school
```

```python
# Plot

plt.figure(figsize=(20,10))
plt.title(label="My Height since Birth")
plt.xlabel(xlabel="My Age (years)")
plt.ylabel(ylabel="My Height in Feet")

#scatterplot
plt.scatter(hdata.index, 
            hdata.values, 
            c="blue", 
            alpha=0.5)

#plot important events
plt.scatter(pa_entrance["x"], pa_entrance["y"], c="red")
plt.scatter(started_running["x"], started_running["y"], c="red")
plt.scatter(middle_schl["x"], middle_schl["y"], c="red")

plt.show()
```

As you might be able to tell, I grew a very small amount since I entered PA; I belive this is due to the fact that my average sleep per night drastically decreased and my eating habits became irregular. Of course, such changes in my growth could have been due to genetics / other factors, but the fact that my growth essentially stopped when I entered PA is alarming.
