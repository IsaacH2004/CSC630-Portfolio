### What about the data should we investigate? 

While I live in Anodver currently, I spent most of my life in Lynnfield, MA. Let's make some visualizations based on data from Lynnfield homes. 

```python
import pywaffle as Waffle # useful visualization tool

lfdata = data[data["CITY"] == "Lynnfield"]
lfdata["ELEMENTARYSCHOOL"].value_counts() # Lynnfield homes & elementary school counts
```

```python
lfdata
```

```python
# the elementary school names are a bit disorganized, let's fix them: 
lfdata.loc[3330,"ELEMENTARYSCHOOL"] = "Huckleberry"
lfdata.loc[2228,"ELEMENTARYSCHOOL"] = "Summer St"
lfdata
```

```python
# Time to plot
fig = plt.figure(FigureClass=Waffle.Waffle, 
                 figsize=(10,10), 
                 values=lfdata["ELEMENTARYSCHOOL"].value_counts(),
                 rows=1)

print("Green: Huckleberry (4)", "Orange: Summer Street (2)", sep="\n") # Interesting! 
```

```python
# For our second plot, we can take a look at the residency types in Lynnfield. 
lfdata["PROPTYPE"].value_counts()
```

```python
# Since all homes are single-family (denoted by "SF"), our pictogram will likely be a bit 
# boring. To remedy this, I'll add a few condominums to the data: 
lf_hometypes = lfdata["PROPTYPE"].value_counts()
lf_hometypes = pd.concat([lf_hometypes, pd.Series(3, index=["CC"])])
lf_hometypes # looks good
```

```python
# Plot
fig = plt.figure(FigureClass=Waffle.Waffle, 
                 figsize=(10,5), 
                 values=lf_hometypes,
                 rows=5,
                 icons=["home"]*2,
                 tight=False)

print("Green: Single-family (7)", "Orange: Condos (3)", sep="\n")
```

### One Last Plot

For this last plot, we should try to take even greater advantage of PyWaffle and Matplotlib (especially because the plots above are not that interesting). We can make a larger, more complex graph to depict selling prices of homes in Boston, MA's largest city. 

Note: I borrow some code for the color and the legend + title from http://sigmaquality.pl/data-plots/perfect-plots-waffle-plot/ 

```python
bos_price
```

```python
bos_price = data[data["CITY"] == "Boston"]
bos_price["SOLDPRICE"] = bos_price["SOLDPRICE"].apply(lambda x: round(x, ndigits = -6)) # round to the nearest million
bos_price.groupby("SOLDPRICE").count()["CITY"]
```

```python
colors2 = [plt.cm.RdYlBu(i/float(10)) for i in range(10)]
colors2
```

```python
# Plot
fig = plt.figure(FigureClass=Waffle.Waffle, 
                 figsize=(10,10), 
                 values=bos_price.groupby("SOLDPRICE").count()["CITY"],
                 rows=20,
                 icons=["home"]*10,
                 tight=False,
                 colors=colors2,
                 plots={
                 '111': {
                 'values': bos_price.groupby("SOLDPRICE").count()["CITY"].values,
                 'labels': ["$0","$1","$2","$3","$4","$6","$7","$9","$10","$15"],
                 'legend': {'loc': 'upper left', 'bbox_to_anchor': (1.0, 1), 'fontsize': 14},
                 'title': {'label': 'Price of Houses in Boston, MA in Millions of USD', 'loc': 'center', 'fontsize':24,'alpha':0.8}
        },
    },)
```

## Takeaways

From the visualization, we can discern that most house selling prices in Boston, MA are either under 500k USD or between 500k and 1,500k. A few sell prices are between 1,500k and 2,500k, while even fewer are between 2,500k and 3,500k. A very, very, very small number of the sell prices are above these ranges, with the highest being around 15,000k.
